## Image processing from the video

#### Since the topics are taught using diagrams and pictures we need to see if the those pictures and accurate with the information and are factually right.

```python
import cv2, os
import subprocess

video_url = "https://www.youtube.com/watch?v=Sr8rRxN2J18"
output_file = "video.mp4"

# Download video using yt-dlp
subprocess.run(["yt-dlp", "-f", "mp4", "-o", output_file, video_url])

```

```python
import cv2
import os
from skimage.metrics import structural_similarity as ssim


import numpy as np

video_path = "video.mp4"
frame_dir = "frames_on_change"
os.makedirs(frame_dir, exist_ok=True)

cap = cv2.VideoCapture(video_path)

ret, prev_frame = cap.read()
count = 0
threshold = 0.90  # SSIM threshold, lower = more sensitive

if ret:
    prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
    cv2.imwrite(f"{frame_dir}/frame_{count:04d}.jpg", prev_frame)
    count += 1

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Calculate Structural Similarity Index (SSIM)
    similarity, _ = ssim(prev_gray, gray, full=True)

    if similarity < threshold:
        cv2.imwrite(f"{frame_dir}/frame_{count:04d}.jpg", frame)
        prev_gray = gray
        count += 1

cap.release()
```
